```python
import numpy
import matplotlib.pyplot as plt

x = numpy.array(range(-200,200))
y = 1/x

ax = plt.gca()

plt.plot(x, y)
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
plt.title('hyperbola')
plt.grid(True)
plt.xlim(-10,10)
plt.show()
```
